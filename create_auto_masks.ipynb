```python
from skimage.filters import threshold_otsu
from skimage.io import imread, imsave
from skimage.color import rgb2gray, rgb2hsv, hsv2rgb
from skimage.morphology import square, closing, opening
from skimage.segmentation import clear_border
import numpy as np
import os
from glob import glob
from tqdm import tqdm
import pandas as pd
from snippets import MaskOverlay
```

```python
luminaux = pd.read_csv('/mnt/data4/tlazard/data/luminaux_BRCA/labels_luminaux.csv', sep=";")
luminaux = luminaux['ID'].values
mo = MaskOverlay(color='b', alpha=0.3)
files = glob('/mnt/data4/gbataillon/DataCurie/downjpg/*.jpg')
len(luminaux)
```

## Creates masks, saves them and their visualisation (luminaux BRCA)

```python
files = glob('/mnt/data4/gbataillon/DataCurie/downjpg/*.jpg')
annot_path = '/mnt/data4/tlazard/data/luminaux_BRCA/annotations'
already_done = os.listdir(os.path.join(annot_path, 'auto_visu'))
already_done = [os.path.splitext(os.path.basename(x))[0] for x in already_done]
for f in files:
    num_histo, _ = os.path.splitext(os.path.basename(f))
    if num_histo in luminaux:
        if num_histo not in already_done:
            im = imread(f)
            im_gray = rgb2gray(im)
            t = threshold_otsu(im_gray)
            mask = opening(closing(im_gray<t, selem=square(15)), selem=square(8))
            overlay_path = os.path.join(annot_path, 'auto_visu', num_histo+'.jpg')
            mask_path = os.path.join(annot_path, 'auto', num_histo+'.npy')
            _ =mo(im=im, mask=mask)
            mo.save_mask(mask_path)
            mo.save_overlay(overlay_path)
            print('processing')
        else:
            print('processed')
```

```python
im = imread('~/Documents/cbio/data/TCGA_TNBC/image_tcga_26.jpg')
im_gray = rgb2gray(im)
t = threshold_otsu(im_gray)
dilated_mask = closing(im_gray<t, selem=square(10))
cleared_mask = clear_border(dilated_mask)


```

```python
#To HSV
alpha = 0.3
rgb_mask = np.zeros((cleared_mask.shape[0], cleared_mask.shape[1], 3))
rgb_mask[:,:,2] = cleared_mask # Masque rouge
im_hsv = rgb2hsv(im)
mask_hsv = rgb2hsv(rgb_mask)

im_hsv[..., 0] = mask_hsv[..., 0]
im_hsv[..., 1] = mask_hsv[..., 1]*alpha

im_masked = hsv2rgb(im_hsv)
```

```python
import matplotlib.pyplot as plt
plt.imshow(im_masked)
imsave('~/Documents/cbio/image_masked.jpg', im_masked)
```
